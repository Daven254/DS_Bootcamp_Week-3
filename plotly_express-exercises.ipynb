#### Task 1 
##### Import the Plotly Express library as px
##### Use the Plotly Express library to pull in the dataset `iris` and save it as df
#### Result:


```python
import plotly.express as px
df = px.data.iris()

df
```

#### Task 2 
##### Create a histogram of petal_length colored by species. Add a title, choose a number of bins, and change the axes labels, and the color palette as you see fit.
#### Result:


```python
fig = px.histogram(
    df, x="petal_length", 
    color="species", 
    hover_data=df.columns, 
    title='Petal Lengths by Species', 
    labels={'petal_length': 'Petal Length'}, 
    nbins=50, 
    color_discrete_sequence=px.colors.sequential.Aggrnyl)

fig.show()
```

#### Task 3 
##### Create a box plot of `sepal_length` colored by species.  Add a title, change axes labels, and the color palette as you see fit.
#### Result:


```python
fig = px.box(df, x = 'sepal_length' , 
             color = 'species', 
             title = 'Sepal Lengths per Species', 
             labels={'sepal_length':'Sepal Length'}) 


fig.show()
```

#### Task 3 
##### Use the code from Task 2 and create a linear regression by group.  Show the model summary for your group of choice.  Is this a good model?
#### Result:


```python
results=px.get_trendline_results(fig)
results
```

```python
query_results = results.query("group=='control'")
```

```python
query_results.px_fit_results.iloc[0].summary()
```

R-squared is really low, so the model cannot really explain the variance of the dependent variable (exp2).

p(t-test) is also not very good, it is too high. This suggests that the independent variable (exp1) is not a strong indicator of model to predict exp2.

p(f-stat) is high as well. This means that the model does not have predictive strength to predict the dependent variable based on the independent variable.

The p(JB) value is high. This implies that the residuals are normally distributed.

#### Task 4
##### Use gapdf and create a choropleth map of `gdpPercap` in year 1982 for the world. 
#### Result:


```python
gapdf_1982 = gapdf.query("year == 1982")
```

```python
fig = px.choropleth(gapdf_1982, locations="iso_alpha",
                    color="lifeExp", # lifeExp is a column of gapminder
                    hover_name="country", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Agsunset_r)
fig.show()
```

#### Task 5
##### Use `eldf` and load in the geoJSON that goes with it to make a map of the votes for Coderre.
#### Result: 


```python
geojson = px.data.election_geojson()
eldf = px.data.election()


fig = px.choropleth(eldf, geojson=geojson, 
                    color="Coderre",
                    locations="district", 
                    featureidkey="properties.district", color_continuous_scale=px.colors.sequential.Agsunset_r,
                    projection="mercator"
                   )

fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
